```python
from os import path
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
text = """
machine learning
data visualization
Python
data visualization
"""
```

```python
responses = text.lower().split("\n")[1:-1]
phrases = [reduce(lambda a,b: a+"_"+b,phrase.split(" ")) for phrase in responses]
wordcloud = WordCloud().generate(reduce(lambda a,b: a+" "+b,phrases))
```

```python
plt.imshow(wordcloud)
plt.axis("off")
```

Rescale for relative counts

```python
wordcloud = WordCloud(max_font_size=40, relative_scaling=.5).generate(reduce(lambda a,b: a+" "+b,phrases))
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
